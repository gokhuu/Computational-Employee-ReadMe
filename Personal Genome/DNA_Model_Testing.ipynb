```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
lst = ['VCFs/cleaned_dna/chr1.csv', 'VCFs/cleaned_dna/chr2.csv','VCFs/cleaned_dna/chr3.csv','VCFs/cleaned_dna/chr4.csv','VCFs/cleaned_dna/chr5.csv',
'VCFs/cleaned_dna/chr6.csv','VCFs/cleaned_dna/chr7.csv','VCFs/cleaned_dna/chr8.csv','VCFs/cleaned_dna/chr9.csv','VCFs/cleaned_dna/chr10.csv',
'VCFs/cleaned_dna/chr11.csv','VCFs/cleaned_dna/chr12.csv','VCFs/cleaned_dna/chr13.csv','VCFs/cleaned_dna/chr14.csv','VCFs/cleaned_dna/chr15.csv',
'VCFs/cleaned_dna/chr16.csv','VCFs/cleaned_dna/chr17.csv','VCFs/cleaned_dna/chr18.csv','VCFs/cleaned_dna/chr19.csv','VCFs/cleaned_dna/chr20.csv',
'VCFs/cleaned_dna/chr21.csv','VCFs/cleaned_dna/chr22.csv','VCFs/cleaned_dna/chrX.csv','VCFs/cleaned_dna/chrY.csv']
df_list = []

for i in lst:
	df = pd.read_csv(i, header=0,index_col=0)
	df_list.append(df)

dna = pd.concat(df_list)

dna = dna.T 
display(dna.head())
```

```python
#encode data
reference_dict = {'Homozygous Reference':0, 'Heterozygous Alternate':1, 
					'Heterozygous Alternate 1/2':2,'Homozygous Alternate':3}

dna = dna.replace(reference_dict)
display(dna.head())
```

```python
target_dict = {'14765x2-1':2, '14763x7-1':2, '14799x1-1':2, '14858x3-1':2,
       '14746x8-1':2, '14824x13-1':2, '14739x3-1':2, '14710x6-1':2, '14781x16-1':2,
       '1601-1':1, '1401-1':1, '902-1':0, '901-1':1, '1001-1':1, 'BXS0110-1':0,
       'BXS0111-1':0, 'BYS0112-1':0, 'BXS0114-1':0, 'BXS0115-1':0, 'BXS0116-1':0,
       'BXS0117-1':0, 'GM23716-1':0, 'GM23720-1':0, 'GM25256-1':0, 'PGP1-1':0}

target = [target_dict[i] for i in dna.index]
target
```

```python
#import library
from sklearn.model_selection import train_test_split

#split training and testing data
x_train, x_test, y_train, y_test = train_test_split(dna, target, test_size=.05, random_state=25)

print(x_test.index,y_test)
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.tree import export_text
from sklearn.metrics import accuracy_score

feature_names=[i for i in dna.columns]

dt = DecisionTreeClassifier(criterion='entropy', splitter='best',random_state=0)
dt.fit(x_train,y_train)
dt_prediction = dt.predict(x_test)

dt_accuracy = accuracy_score(y_test, dt_prediction)
print('Model Accuracy: {}'.format(dt_accuracy*100))
plot_tree(dt,filled=True)
print(export_text(dt, feature_names=feature_names))
```
