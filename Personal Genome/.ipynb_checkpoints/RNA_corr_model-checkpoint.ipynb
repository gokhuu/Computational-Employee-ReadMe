# RNA Model trimmed data

```python
#import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('fpkm_corr.csv', header=0,index_col=0)
df.head()
```

```python
from sklearn.model_selection import train_test_split

y = df.Autism
x = df.iloc[:,1:]

x_train, x_test, y_train,y_test = train_test_split(x,y,test_size = .1, random_state=50)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score
from random import *

rand_state_nn = []
nn_acc = []
for j in range(100):

    random_state = randint(0,1000)
    rand_state_nn.append(random_state)
    nn = MLPClassifier(alpha=0.0001,max_iter=500, random_state=random_state)
    nn.fit(x_train,y_train)
    nn_prediction= nn.predict(x_test)

    nn_accuracy = accuracy_score(y_test, nn_prediction)
    nn_acc.append(nn_accuracy)

    #print('Model Accuracy: {}'.format(nn_accuracy*100))
    plt.plot(nn.loss_curve_)
    
print(np.mean(nn_acc)*100)
```

```python
from sklearn.ensemble import RandomForestClassifier
rand_state_rf = []
rf_acc = []
for j in range(100):
    random_state = randint(0,1000)
    rand_state_rf.append(random_state)
    
    rf = RandomForestClassifier(n_estimators=500, random_state=random_state)
    rf.fit(x_train, y_train)
    rf_prediction = rf.predict(x_test)
    rf_accuracy = accuracy_score(y_test, rf_prediction)
    rf_acc.append(rf_accuracy)
    #print('Model Accuracy: {}'.format(rf_accuracy*100))

print(np.mean(rf_acc)*100)
```
