# RNA and DNA Model

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rna = pd.read_csv('Organoid Files/all_autism_fpkm_iqr_1.5_outliers_multi_class.csv', header=0,index_col=0)
dna = pd.read_csv('DNA_csv/dna_table_edit_t_mutli_replicates.csv', header=0,index_col=0)
display(rna.head())
display(dna.head())
```

```python
#label encoding
from sklearn.preprocessing import LabelEncoder
cat_features = [i for i in dna.columns]
encoder = LabelEncoder()
dna_encoded = dna[cat_features].apply(encoder.fit_transform)
dna_encoded
```

```python
comb = rna.merge(dna_encoded, how='outer', left_index=True, right_index=True)
comb
```

```python
cat_features.append('Autism')
for i in cat_features:
    comb[i] = pd.Categorical(comb[i])

y = comb.Autism
x = comb.iloc[:,1:]
x.shape
```
