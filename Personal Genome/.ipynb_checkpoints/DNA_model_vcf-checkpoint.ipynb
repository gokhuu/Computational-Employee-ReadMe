```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('DNA_csv/dna_table_edit_t.csv', header=0, index_col = 0,low_memory=False)
df.head()
```

```python
y = df.Autism
x = df.iloc[:,1:]
```

```python
cols = [i for i in x.columns]
for i in cols:
    x[i] = x[i].astype('category')
x.head()
```

```python
from tensorflow import keras
from tensorflow.keras import layers

#create early stop
early_stop = keras.callbacks.EarlyStopping(monitor='loss',patience=2, verbose=1)

#create model
model = keras.Sequential([
    layers.Dense(units=100, activation='relu',input_shape=[8613]),
    layers.BatchNormalization(),
    layers.Dropout(0.2),
    layers.Dense(units=1,activation='sigmoid')
])

#compile model with functions
model.compile(
    optimizer='rmsprop',
    loss='binary_crossentropy',
    metrics=['accuracy']
)
```
